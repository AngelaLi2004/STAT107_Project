# Section 5: Data Science — Predicting Early-Career Outcomes Using Machine Learning
In Section 4, I proposed exploring how well early-career outcomes—such as Job Offers and Starting Salary—can be predicted from students’ academic performance, practical experience, and soft skills. This leads to the following question:

Research Question:

Can we predict a student's number of Job Offers using features such as University GPA, internships, projects, soft skills, and networking?

To answer this, I use a supervised machine learning model introduced in Module 5:

- K-Nearest Neighbors (KNN) Regression

This model predicts numerical outcomes by examining the most similar observations (neighbors) in the dataset.



```python
# import libraries
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import numpy as pd
import pandas as pd

```

```python
# Create an Instance of the Model
model_knn = KNeighborsRegressor(n_neighbors=5)
```

```python
# Train the Model Using a 70/30 Train-Test Split
# Load dataset
df = pd.read_csv("education_career_success.csv")

# Select features to predict Job Offers
features = [
    "University_GPA",
    "Internships_Completed",
    "Projects_Completed",
    "Soft_Skills_Score",
    "Networking_Score"
]

X = df[features]
y = df["Job_Offers"]

# 70% / 30% split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.30, random_state=42
)

# Train the model
model_knn.fit(X_train, y_train)
```

```python
# Predict Job Offers and Evaluate Model Performance
y_pred = model_knn.predict(X_test)

# Compute model accuracy metrics
mse = mean_squared_error(y_test, y_pred)
rmse = mse ** 0.5
r2 = r2_score(y_test, y_pred)

rmse, r2

```

```python
y_pred
```

The KNN regression model performed extremely well in predicting Job Offers. The RMSE of approximately 0.096 indicates that the model’s average prediction error is less than one-tenth of a job offer, which is remarkably small. Additionally, the R² score of 0.995 shows that 99.5% of the variation in Job Offers can be explained by the five input features (GPA, internships, projects, soft skills, and networking).

# Section 6: A Different Data Visualization
In this section, I create a different type of visualization from the ones used in Section 3. While Section 3 included a scatter plot, histogram, and bar chart, here I use a box plot to analyze variation in starting salaries across different academic fields.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

df.boxplot(
    column="Starting_Salary",
    by="Field_of_Study",
    grid=False
)

plt.title("Distribution of Starting Salary by Field of Study")
plt.suptitle("")   # Removes default pandas title
plt.xlabel("Field of Study")
plt.ylabel("Starting Salary")
plt.xticks(rotation=45)

plt.show()
```

This box plot illustrates how starting salaries differ substantially across academic majors. Fields such as Medicine, Computer Science, and Engineering show both high median salaries and wide salary ranges, indicating strong labor-market demand as well as considerable variability in individual earning outcomes. In contrast, majors like Education, Psychology, and Arts have noticeably lower and more tightly clustered salary distributions, suggesting more limited earning potential and less variation among graduates. Meanwhile, fields such as Finance, Business, and Marketing occupy a middle range with moderate medians and several high-earning outliers. Overall, the visualization demonstrates that early-career salary outcomes are strongly shaped by a student’s field of study, with some majors providing not only higher earnings but also a greater spread of opportunities.

```python
plt.figure(figsize=(12, 6))

df.boxplot(
    column="Job_Offers",
    by="Field_of_Study",
    grid=False
)

plt.title("Distribution of Job Offers by Field of Study")
plt.suptitle("")   # Removes automatic Pandas boxplot title
plt.xlabel("Field of Study")
plt.ylabel("Job Offers Received")
plt.xticks(rotation=45)

plt.show()
```

This box plot shows how the number of job offers varies across different academic fields. Several majors—such as Computer Science, Engineering, Finance, and Medicine—have higher medians and broader interquartile ranges, indicating both stronger hiring demand and more variability in outcomes. In contrast, fields such as Arts, Education, and Psychology show lower medians and tighter distributions, suggesting more limited job opportunities and less variation among students. The presence of outliers in some fields also highlights that exceptional cases exist even in majors with lower overall medians.
